```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-1, 2, 100)
plt.figure()
```

```python
plt.scatter(x,np.exp(x), label="EXP INC")

plt.scatter(x,np.exp(-x), label="EXP DEC")

plt.scatter(x, x, label="LINEAR")

plt.xlabel('$x$')

plt.ylabel('$y$')

plt.legend(loc=0)

plt.title("Scatter Plot")
plt.show()
```

```python
"""Draw Sine wave using Dashed line"""

x1 = np.arange(0, 2*(np.pi), 0.1)

y1 = np.sin(x1)

plt.plot(x1, y1, color='green', linestyle='dashed')

x_coordinates = [0, 6.2]

y_coordinates = [0.5, 0.5]

plt.plot(x_coordinates, y_coordinates)

plt.title("Dashed Sine wave")

plt.show()
```

```python
# Squares Wave
x3 = np.arange(0, 3*(np.pi), 0.1)

y3 = np.square(x3)

plt.plot(x3, y3)

plt.xlabel('Time')

plt.ylabel('Amplitude')

plt.title("Squre wave")

plt.show()
```

```python
# Trigular Wave

from scipy import signal

timePoints = np.linspace(0, 1, 500)

plt.plot(timePoints, signal.sawtooth(2* np.pi*5*timePoints))

plt.xlabel('Time')

plt.ylabel('Amplitude')

plt.axhline(y = 0, color='k')

plt.ylim(-2, 2)

plt.title("Trigular wave")

plt.show()
```

```python
# Sine Wave

x4 = np.arange(0,2*(np.pi), 0.1)

y4 = np.sin(x4)

plt.plot(x4, y4)

plt.xlabel('Time')

plt.ylabel('Amplitude')

plt.title("Sine wave")

plt.show()
```

```python
#Cos wave

x5 = np.arange(0, 2*(np.pi), 0.1)

y5 = np.cos(x5)

plt.xlabel('Time')

plt.ylabel('Amplitude')

plt.plot(x5, y5)

plt.title("Cos wave")

plt.show()
```
